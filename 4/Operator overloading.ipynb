Preorder traversal: root first then left till the down and then go to right
    

Postorder travesal: last nodes first then go back to the root


```python
#include <iostream>
using namespace std;


int a = 2;
int b = 1;
int *a_ptr = &a , *b_ptr = &b;
    
cout << a << endl;
cout << b << endl;
cout << a_ptr << endl;
cout << b_ptr << endl;

if (a_ptr < b_ptr){
    cout << "no";
}

if (a < b){
    cout << "yes";
}
```
